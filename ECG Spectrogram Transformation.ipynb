```python
import wfdb as wf
data = wf.rdsamp('/home/hsiehch/dataset/' + 'Record_1')
data = data[0]
data.shape = (data.shape[0], )
```

```python
import pylab as plt
plt.figure(figsize=(16,5))
plt.xlim(0,3000)
plt.plot(data)
plt.show()
```

```python
from scipy.signal import spectrogram
import numpy as np

f, t, Sx = spectrogram(data[:3000], fs=300, nperseg=300, noverlap=150, scaling='spectrum', detrend=False, window='hanning')
Sx = np.log10(Sx)
plt.figure(figsize=(16,5))
plt.pcolormesh(t, f, Sx)
plt.show()
```
